Model making using ANN

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
x_train = np.array(x_train).astype(np.float32)
y_train = np.array(y_train).astype(np.float32)
y_train = y_train.reshape(-1) 

model = keras.Sequential([
    keras.layers.Input(shape=(26,)),
    keras.layers.Dense(26, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp = model.predict(x_test)
```

```python
yp[:5]
```

```python
y_pred =[]
for y in yp:
    if y >0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test,y_pred))
```

Predictions

```python
def check_telecom_churn(arr):
    for index, value in enumerate(arr):
        if value == 1:
            print(f"Person {index + 1} will churn the telecom.")
        elif value == 0:
            print(f"Person {index + 1} will stay with the telecom.")
        else:
            print(f"Invalid value at index {index + 1}. Please provide only 0 or 1.")
check_telecom_churn(y_pred[:10])
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
